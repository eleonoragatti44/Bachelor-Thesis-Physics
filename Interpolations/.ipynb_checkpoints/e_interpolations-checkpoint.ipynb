# Preparazione Dati

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy.optimize import curve_fit
import functions as f
```

Importo file e aggiungo una colonna _check_ che alterna True e False

```python
data = pd.read_csv('data_fs.dat', delim_whitespace=True, decimal=",")
f.add_bool(data)
```

```python
data_int = data[data['check'] == True]
data_check = data[data['check'] == False]
```

"Filtro" i dati rimuovendo i punti sul bordo, ovvero quelli con |x| o |y| > 0.3

```python
mask = (abs(data_check["x"]) < 0.3) & (abs(data_check["y"]) < 0.3)
data_mask = data_check[mask]
```

---------------

# Interpolazione

L'interpolazione dei dati sarà effettuata su sul dataset `data_int` e la bontà dei risultati sarà verificata valutando la differenza tra dato vero appartenete al dataset `data_check` e quello ottenuto tramite la funzione interpolante.

Seguiremo due approci alternativi, che riguardano:
- il metodo `interp2d` del modulo `scipy.interpolate`
- il metodo `curve_fit` del modulo `scipy.optimize`

Per completezza è stato utilizzato anche il metodo `griddata`

-------

## Interp2d


### Ellitticità

```python
f_int_e = f.int_lin(data_int, "e")
f_check_e = f.int_lin(data_check, "e")

d_e = f.delta(data_check, f_int_e, f_check_e)
d_e_mask = f.delta(data_mask, f_int_e, f_check_e)
```

#### Descrizione dei grafici
- Errore assoluto tra dati ricavati dall'interpolazione e dati esatti
- Errore assoluto tra dati ricavati dall'interpolazione e dati esatti espresso in scala logaritmica
- Contour plot della differenza in scala logaritmica

```python
f.graph(data_check, data_int, d_e, "ellitticità")
```

Si nota che i dati con un errore maggiore sono quelli sul bordo, proviamo perciò a escludere i punti di bordo in fase di verifica, avendoli però considerati in fase di interpolazione.

```python
f.graph(data_mask, data_int, d_e_mask, "ellitticità mask")
```

-------

## Curve Fit

### Ellitticità

```python
X = [data["x"], data["y"]]
X_check = [data_check["x"], data_check["y"]]
X_check_mask = [data_mask["x"], data_mask["y"]]
X_int = [data_int["x"], data_int["y"]]
```

```python
popt, pcov = curve_fit(f.func, X_int, data_int["e"])
test_func_e = f.func(X_check, *popt)

popt1, pcov1 = curve_fit(f.func, X, data["e"])
test_func_e2 = f.func(X_check, *popt1)
```

Il seguente grafico confronta l'errore assoluto ottenuto tramite un fit eseguito sui dati completi (in blu) o su metà dei dati (in rosso).

```python
#abs(test_func_e2-data_check["e"]).plot(color="blue", label="Fit complete data")
abs(test_func_e-data_check["e"]).plot(color="firebrick", label="curve_fit")
plt.title("Ellipticity")
plt.legend(loc='upper left')
plt.xlabel("Index")
plt.ylabel("Error")
plt.savefig('./Plots/Error_curve_fit.png', dpi=600)
```

Confronto l'errore con quello relativo al metodo di interpolazione `interp2d` (in verde). Sono esclusi i punti di bordo poichè su questi l'errore relativo alla funzione ottenuta tramite `interp2d` risulta essere molto superiore a quello relativo a `curve_fit`, in tali punti l'errore non è comparabile.

```python
abs(test_func_e-data_check["e"]).plot(color="blue", label="curve_fit")
plt.plot(data_check.index, d_e, color = "green", label="interp2d")
plt.legend(loc='upper left')
plt.xlabel("Index")
plt.ylabel("Error")
plt.title("Ellipticity")
plt.savefig('./Plots/ErrorComparisonAll', dpi=600)
```

```python
#Plot da inserire in tesi

fig = plt.figure(figsize(16,8))
ax1 = fig.add_sublot(121)
ax1.plot(data_check.index, d_e, color = "green", label="interp2d")
ax1.set_xlabel("Index")
ax1.set_ylabel("Error")
ax1.set_title("Ellipticity")
plt.legend(loc='upper left')

ax2 = fig.add_sublot(122)
pt_int = ax2.scatter(data_int["x"], data_int["y"], color="red")
pt_check = ax2.scatter(data_check["x"], data_check["y"], color="blue")
plt.legend((pt_int, pt_check), ('Interpolation', 'Check'), scatterpoints=1, loc='lower right', fontsize=10)
```

```python
#abs(test_func_e2-data_check["e"]).plot(color="blue", label="Fit complete data")
abs(test_func_e-data_check["e"]).plot(color="blue", label="curve_fit")
plt.plot(data_mask.index, d_e_mask, color = "green", label="interp2d")
plt.legend(loc='upper left')
plt.xlabel("Index")
plt.ylabel("Error")
plt.title("Ellipticity")
plt.savefig('./Plots/ErrorComparison', dpi=1200)
```

Ci sono chiaramente alcuni punti problematici per quanto riguarda il metodo di interpolazione `interp2d`

Scelgo una riga di punti e plotto il parametro relativo a quella riga

```python
f.check_graph_row(data, f_int_e, test_func_e, test_func_e2, 'e')
```

Scelgo una colonna di punti e plotto il parametro relativo a quella colonna

```python
f.check_graph_col(data, f_int_e, test_func_e, test_func_e2, 'e')
```

```python
i = 0
step = abs(data["y"].max()-data["y"].min())/13
for i in range(6):
    
    
    
    lim_inf = data_check["y"] <= data_check["y"].max() - i*step
    lim_sup = data_check["y"] >= data_check["y"].max() - step - i*step
    data_test = data_check[lim_inf & lim_sup]
    ellit = f.val_int(data_test, f_int_e)
    
    X_check = [data_test["x"], data_test["y"]]
    curvefit = f.func(X_check, *popt)
    
    fig = plt.figure(figsize=(24,10))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.scatter(data["x"], data["y"], color="orange")
    ax1.scatter(data_test["x"], data_test["y"], color="blue")
    plt.xlabel("x [m]")
    plt.ylabel("y [m]")
    ax2 = fig.add_subplot(1, 2, 2)
    ax2.scatter(data_test.index, data_test["e"], color="red", label="e true")
    ax2.plot(curvefit, label="curve_fit", color="blue", marker="+", markersize=14)
    #ax2.plot(test_func2, label="curve_fit", color="blue")
    ax2.plot(data_test.index, ellit, label="interp2d", color="green", marker="x", markersize=14)
    plt.legend()
    plt.xlabel("Index")
    plt.ylabel("Ellipticity")
    plt.rcParams.update({'font.size': 20})
    fig.savefig(f'./Plot_e/ProvaRow_{i}', dpi=400)
    plt.close(fig)
```

I grafici si trovano nella cartella "Plot". Per ogni riga e per ogni colonna è stato fatto il confronto tra ellitticità vera e "misurata".  

**Righe:** si nota che `interp2d` funziona bene tranne che per le righe 0, 2, 10, 12 (la 0 corrisponde alla riga più in alto e la 12 a quella più in basso)

**Colonne:** in questo caso non è stata plottata la curva ottenuta con curve_fit poichè rappresentata su scala troppo grande (problema risolvibile) --> **RISOLTO**  
Si nota che una riga si e una no è presente un punto iniziale della spezzata `interp2d` che crea problemi, al momento non so risolvere e giustificare tale comportamento.  
Facendo riferimento al grafico qui sopra si nota che i punti problematici sono 6, e facendo un confronto tra i grafici riferiti alle righe e alle colonne (e analizzando il contour plot) si nota che è la prima riga dall'alto è la responsabile di tutto.

Rimuovo dal dataset la riga problematica

```python
data_noprob = data[data["y"]<0.3]
data_prob = data[data["y"]>0.3]

ellit_noprob = f.val_int(data_noprob, f_int_e)

fig = plt.figure(figsize=(12,8))
plt.scatter(data.index, data["e"], marker='o', color="orange", label="Ellitticità vera")
plt.scatter(data_prob.index, data_prob["e"], marker='x', color="red", label="Punti problematici")
plt.plot(test_func_e, label="Curve Fit", color="blue")
plt.plot(data_noprob.index, ellit_noprob, label="Interp2d", color="green")
plt.legend()
plt.title('Ellipticity Value')
plt.savefig('./Plots/EllipticityComparisonMask', dpi=600)
```
