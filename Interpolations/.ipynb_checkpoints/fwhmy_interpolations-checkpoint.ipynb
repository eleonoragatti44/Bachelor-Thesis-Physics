# fwhm_y

# Preparazione Dati

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy.optimize import curve_fit
import functions as f
```

Importo file e aggiungo una colonna _check_ che alterna True e False

```python
data = pd.read_csv('data_fs.dat', delim_whitespace=True, decimal=",")
f.add_bool(data)
```

```python
data_int = data[data['check'] == True]
data_check = data[data['check'] == False]
```

"Filtro" i dati rimuovendo i punti sul bordo, ovvero quelli con |x| o |y| > 0.3

```python
mask = (abs(data_check["x"]) < 0.3) & (abs(data_check["y"]) < 0.3)
data_mask = data_check[mask]
```

---------------

# Interpolazione

-------

## Interp2d


### full width half maximum (x)

```python
f_int_fwhmy = f.int_lin(data_int, "fwhm_y")
f_check_fwhmy = f.int_lin(data_check, "fwhm_y")

d_fwhmy = f.delta(data_check, f_int_fwhmy, f_check_fwhmy)
d_fwhmy_mask = f.delta(data_mask, f_int_fwhmy, f_check_fwhmy)
```

#### Descrizione dei grafici
- Errore assoluto tra dati ricavati dall'interpolazione e dati esatti
- Errore assoluto tra dati ricavati dall'interpolazione e dati esatti espresso in scala logaritmica
- Contour plot della differenza in scala logaritmica

```python
f.graph(data_check, data_int, d_fwhmy, "fwhm (y)")
```

Si nota che i dati con un errore maggiore sono quelli sul bordo, proviamo perci a escludere i punti di bordo in fase di verifica, avendoli per considerati in fase di interpolazione.

```python
f.graph(data_mask, data_int, d_fwhmy_mask, "fwhm (y)")
```

-------

## Curve Fit

### full width half maximum (x)

```python
X = [data["x"], data["y"]]
X_check = [data_check["x"], data_check["y"]]
X_check_mask = [data_mask["x"], data_mask["y"]]
X_int = [data_int["x"], data_int["y"]]
```

```python
popt, pcov = curve_fit(f.func, X_int, data_int["fwhm_y"])
test_func_fwhmy = f.func(X_check, *popt)

popt1, pcov1 = curve_fit(f.func, X, data["fwhm_y"])
test_func_fwhmy2 = f.func(X_check, *popt1)
```

Il seguente grafico confronta l'errore assoluto ottenuto tramite un fit eseguito sui dati completi (in blu) o su met dei dati (in rosso).

```python
abs(test_func_fwhmy2-data_check["fwhm_y"]).plot(color="blue", label="Fit complete data")
abs(test_func_fwhmy-data_check["fwhm_y"]).plot(color="red", label="Fit half data")
plt.legend(loc='upper left')
plt.xlabel("Index")
plt.ylabel("Error")
```

Confronto l'errore con quello relativo al metodo di interpolazione `interp2d` (in verde). Sono esclusi i punti di bordo poich su questi l'errore relativo alla funzione ottenuta tramite `interp2d` risulta essere molto superiore a quello relativo a `curve_fit`, in tali punti l'errore non  comparabile.

```python
abs(test_func_fwhmy2-data_check["fwhm_y"]).plot(color="blue", label="Fit complete data")
abs(test_func_fwhmy-data_check["fwhm_y"]).plot(color="red", label="Fit half data")
plt.plot(data_mask.index, d_fwhmy_mask, color = "green", label="Interp2d")
plt.legend(loc='upper left')
plt.xlabel("Index")
plt.ylabel("Error")
```
